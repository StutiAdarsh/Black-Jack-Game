```python
import random

suits = ('Hands', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10,
         'King':10, 'Ace':11}

playing1 = True
```

```python
class Cards:
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank+ " of " +self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Cards(suit, rank))
                
    def __str__(self):
        deck_comp=''
        for card in self.deck:
            deck_comp += '\n'+card.__str__()
        return f'The deck has: {deck_comp}' 
        
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self, card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1
            
    def adjust_for_aces(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:
    
    def __init__(self):
        self.total= 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
#--------------------------FUNCTIONS----------------------------------------

def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input('How much chips woud you like  to bet?: '))
        except:
            print('Oops Sorry !! Kindly  enter the number of chips out of 100 total. Enter again \n')
        else:
            if chips.bet > chips.total:
                print(f'Sorry !! The bet amount cannot exceed the total of {chips.total}')
            else:
                break
```

```python
def hit(deck,hand):
    
    hand.add_card(deck.deal())
    hand.adjust_for_aces()
```

```python
def hit_or_stand(deck, hand):
    global playing1
    
    while True:
        
        x = input('Hit or Stand ? Enter h or s: ')

        if x[0].lower() == 'h':
            hit(deck, hand)

        elif x[0].lower() == 's':
            print("Player stands..\nDealer's turn..!!")
            playing1 = False

        else:
            print('Wrong input !! Enter h or s:  ') 
            continue
            
        break
```

```python
def show_some(dealer, player):
    print("\nDealer's hand: \n")
    print("<card hidden>")
    print(dealer.cards[1])
    print("\nPlayer's hand: \n",  *player.cards, sep='\n')

def show_all(dealer, player):
    print("\nDealer's hand: \n",  *dealer.cards, sep='\n')
    print("\nDealer's hand = " ,dealer.value)
    print("\nPlayer's hand: \n",  *player.cards, sep='\n')
    print("\nPlayer's hand = " ,player.value)
```

```python
def dealer_busts(dealer, player, chips):
    print('Dealer busts !!\nPlayer wins..')
    chips.win_bet()
    
def dealer_wins(dealer, player, chips):
    print('Dealer wins !!\nPlayer loses..')
    chips.lose_bet()
    
def player_busts(dealer, player, chips):
    print('Player busts !!\nDealer wins..')
    chips.lose_bet()
    
def player_wins(dealer, player, chips):
    print('Player wins !!\nDealer loses..')
    chips.win_bet()
    
def push(dealer, player):
    print("Dealer and player TIE !!!..Its a push")
```
